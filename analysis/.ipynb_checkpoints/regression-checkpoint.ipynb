```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
%matplotlib inline
%load_ext autoreload
%autoreload 2


from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge 

plt.style.use('ggplot')
```

```python
data = pd.read_csv('cleaned_data.csv')
cols = ['AMAT_fac','MH_fac',
       'Med_AMAT_fac', 'Med_MH_fac', 'Med_SA_fac', 'Med_SMAT_fac',
       'Med_TMAT_fac', 'SA_fac', 'SMAT_fac', 'TMAT_fac','HIVprevalence']
X = data[cols]
y = data['HIVincidence']
```

```python
# Split into train, test data
X_train, X_test, y_train, y_test = train_test_split(X.values, y.values)

# Create dictionary to keep track of metrics
metrics = {}
```

```python
# Look at OLS summary for important parameters
import statsmodels.api as sm
model = sm.OLS(y_train,X_train).fit()
model.summary()
```

```python
# Create DataFrame with p-values of parameters
```

```python
# Predict and calculate RMSE
y_hat = model.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_hat))

metrics['OLS RMSE'] = rmse
```

```python
# Regular Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)
y_hat = model.predict(X_test)

# Metrics
rmse = np.sqrt(mean_squared_error(y_test, y_hat))
r_sq = model.score(X_test, y_test)
residuals = y_test - y_hat

print(rmse)
metrics['Linear Regression'] = rmse

# Plot residuals
plt.plot(y_test-y_hat, 'bo')
```

## Regularization

```python
# Model with regularization requires standardization
from src import helper
from sklearn.linear_model import Ridge

# Get train and test errors at different alphas
ridge_alphas = np.logspace(-2, 4, num=250)
train_errors, test_errors = helper.train_at_various_alphas(X_train, y_train, Ridge, ridge_alphas);
```

```python
# Plot values
tr = train_errors.mean(axis=0)
ts = test_errors.mean(axis=0)

fig, ax = plt.subplots()
ax.plot(tr, 'b--', label="Training Data Errors")
ax.plot(ts, 'g--', label="Testing Data Errors")
ax.legend()
ax.set_xscale('log')
ax.set_xlabel('alpha')
ax.set_ylabel('errors')
```

```python
# Find minimim testing mse errors and use that for alpha
min_mse = min(ts)
idx = np.argmin(ts)
alpha_best = ridge_alphas[idx]
alpha_best, min_mse
print(min_mse, idx)
metrics['Regularization + KFold'] = np.sqrt(min_mse)
```
